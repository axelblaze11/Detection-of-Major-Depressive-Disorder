```python
import numpy as np  
import pandas as pd  
```

```python
#importing Dataset
dataset = pd.read_excel(r'F:\Project\Machine Learning & Artificial Intelligence\Project 1-Major Depressive Disorder\Dataset\delta_pow.xlsx')
dataset.head()
```

```python
#seeing the charts
X=dataset.iloc[:,0:19]
X.head()
```

```python
y=dataset.iloc[:,19]
y.head()
```

```python
#splitting Dataset into train & test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0) 
```

```python
#applying KNN classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier=KNeighborsClassifier(n_neighbors=9)
classifier
```

```python
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
a=cm[0][0]
b=cm[0][1]
c=cm[1][0]
d=cm[1][1]
#calculation Error,Accuracy,Sensitivity and Prediction
Err=(b+c)/(a+b+c+d)
Acc=(a+d)/(a+b+c+d)
SN=a/(a+c)
Prec=a/(a+b)
print(cm)
print('Accuracy=',Acc*100,'%')
print('Error=',Err*100,'%')
print('Sensitivity=',SN*100,'%')
print('Prediction=',Prec*100,'%')
```
